```python
#Data cleaning
#find the null values in the dataset. if there is any null value in any column, 
#then fill it with the mean value of that column.

car.isnull().sum()
```

```python
car["Cylinders"].fillna(car["Cylinders"].mean(), inplace = True)
```

```python
#Value Counts
#what are the different type of Make are there in a dataset, And, what is the count (occurence) of the each Make in the data?
car.head(2)
```

```python
car["Make"].value_counts()
```

```python
#Filtering
#show the records where the origin is Asia or Europe.
car[car["Origin"].isin(["Asia" , "Europe"])]
```

```python
#Removing unwanted records
#Remove the records(rows) where the weight is above 4000.
car[~(car["Weight"]>4000)]
```

```python
#Applying function on column
#increase all the value of 'MPG_City' column by 3
```

```python
car["MPG_City"] = car["MPG_City"].apply(lambda x:x+3)
```
